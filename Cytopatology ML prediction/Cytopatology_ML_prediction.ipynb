```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
dataset=dataset.drop('Unnamed: 32',axis=1)
```

```python
dataset['diagnosis'].unique()

dataset['diagnosis'].value_counts()
```

```python

df=pd.get_dummies(dataset,columns=['diagnosis'],drop_first=True,dtype='int')
df
```

```python
df.info()
```

```python
x=df.iloc[:,df.columns!='diagnosis_M']
x
```

```python
y=df['diagnosis_M']
y
```

### Normalising the Data for better accuracy

```python

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_scaled=sc.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=55)
```

```python
from sklearn import tree
```

```python
Dtree=tree.DecisionTreeClassifier()
```

```python
# Bagging Technique

from sklearn.ensemble import BaggingClassifier
```

```python
bag_clf=BaggingClassifier(estimator=Dtree,n_estimators=50,bootstrap=True,n_jobs=-1,random_state=55)
```

```python
bag_clf.fit(x_train,y_train)
```

```python
train_predict=bag_clf.predict(x_train)
test_predict=bag_clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
cm_train=confusion_matrix(y_train,train_predict)
cm_train
sns.heatmap(cm_train,annot=True,fmt='d')
```

```python
cm_test=confusion_matrix(y_test,test_predict)
cm_test
sns.heatmap(cm_test,annot=True,fmt='d')
```

```python
print('Training accuracy score:',accuracy_score(y_train,train_predict))
print('Testing accuracy score:', accuracy_score(y_test,test_predict))
```
